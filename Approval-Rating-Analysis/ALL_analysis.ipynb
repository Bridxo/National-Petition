```python
def find_same_any(test, wlist):
    a_list = list(test)
    del a_list[a_list.index('alls')]
    del a_list[a_list.index('keyword')]
    
    for i in wlist:
        del a_list[a_list.index(str(i))]
    
    rdf = test[test["35"]==43]
    for _, row in test.iterrows():
        k = 0
        for tw in a_list:
            if(row[str(tw)]):
                k=1
        if not(k):
            rdf = rdf.append(row)
    return rdf
```

```python
wlist_m = list(rdf[rdf["차이"]<-2]["대통령 주"])
wlist_m
```

```python
t_n = find_same_any(test, wlist_m)
```

```python
t_n_m = [str(i) for i in wlist_m]
t_n_m.append("keyword")
t_n_m.append("alls")
```

```python
t_nn = t_n[t_n_m]
t_nn
```

```python
test_b = t_nn.astype('bool')
test_b = test_b.astype(float)
test_b["keyword"] = t_nn["keyword"]
test_b["alls"]= test_b.sum(axis = 1, skipna = True)
test_b
```

```python
t_nn["binary"] = test_b["alls"] - 1
t_nn
```

```python
t_nn["value"] = t_nn["binary"] * t_nn["binary"] * t_nn["alls"]
t_nn
```

```python
t_nn.sort_values(by=['value'],ascending=False)
```
