## 숫자가 포함된 문자열을 받으면 숫자로 바꾸기

```python
def find_int(count):
    i = (re.findall(r'\d+', count))
    ints = int(''.join(map(str, i)))
    return ints
```

## map을 이용하여 전체 변경

```python
df["count"] = df["count"].map(find_int)
df
```
