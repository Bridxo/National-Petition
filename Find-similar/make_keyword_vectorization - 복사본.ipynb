```python
all_set = []
f = open("keyword.txt", 'r',encoding="UTF8")
while True:
    line = f.readline()
    if not line: break
    all_set.append(line[:-1])
f.close()
```

```python
all_set
```

```python
df_name = list(df1["title"])
```

```python
conut_lists = []
for i in range(len(df_name)):
    conut_list = []
    for one_set in all_set:
        conut_list.append(df_name[i].count(one_set))
    conut_lists.append(conut_list)
```
