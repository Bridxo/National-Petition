```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import *
import pandas as pd
import platform
from collections import Counter
from matplotlib import font_manager, rc
from konlpy.tag import Twitter; t = Twitter()
import numpy as np
```

```python
df1 = pd.read_csv("../petition_data_all.csv")
```

```python
df1 = df1[df1["count"]>=100000]
df1
```

```python
nlist = list(df1["num"])
nlist
```

```python
ko_con_text = list(df1["petition_overview"])
ko_con_text = ' '.join(ko_con_text)
print(len(ko_con_text))
```

```python
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
tokens_ko = list(set(tokens_ko))
tokens_ko
```

```python
all_set = tokens_ko
len(all_set)
```

```python
all_set
```

```python
df_name = list(df1["petition_overview"])
```

```python
conut_lists = []
for i in range(len(df_name)):
    conut_list = []
    for one_set in all_set:
        conut_list.append(df_name[i].count(one_set))
    conut_lists.append(conut_list)
```

```python
len(conut_lists)
```

```python
len(all_set), len(set(all_set)), len(conut_lists[0])
```

```python
conut_lists[0]
```

```python
lists = [{'key_word' : "key_word", str(i): 0} 
         for i in nlist]
test = pd.DataFrame(lists)

for i in range(len(list(df1["num"]))):
    test = test.drop(i,0)
    
test
```

```python
tlist = []
for j in range(len(conut_lists)):
    tlist.append(nlist[j])
tlist.append(all_set[i])
test.loc[0] = tlist
test
```

```python
for i in range(len(all_set)):
    tlist = []
    for j in range(len(conut_lists)):
        tlist.append(conut_lists[j][i])
    tlist.append(all_set[i])
    test.loc[i] = tlist
    
test
```
