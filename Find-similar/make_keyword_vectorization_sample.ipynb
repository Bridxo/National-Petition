```python
def make_df_set_list(df,typ="list"):
    if typ == "set":
        df_list = list(set(df["Source"]))
        df_list.extend(list(set(df["Target"])))
        df_list = list(set(df_list))
    else:
        df_list = list(df["Source"])
        df_list.extend(list(df["Target"]))
    return df_list
```

```python
df_name = [df1,df2,df3,df4]
```

```python
df_set_list = []
for df in df_name:
    df_set_list.append(make_df_set_list(df, "set"))
    
all_set = list(set(sum(df_set_list, [])))
```
