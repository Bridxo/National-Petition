## 월별

```python
names = list(set(df["category"]))
names.append("all")
```

```python
def fied_names_num(df, names):
    lists = []
    for name in names:
        if(name=="all"):
            break
        print(name, " : ", len(df[df["category"]==name]["category"]))
        lists.append(len(df[df["category"]==name]["category"]))
    print("all : ",len(df["category"]))
    lists.append(len(df["category"]))
    return lists
```

```python
i= 0
fied_names_num(df[df["count"]>=i],names)
```

## 월별

```python
def find_day(count):
    return count[:-3]
```

```python
df["start-days"] = df["start-days"].map(find_day)
df
```

```python
names = list(set(df["start-days"]))
names.append("all")
```

```python
names.sort()
names
```

```python
def fied_names_num(df, days):
    lists = []
    for day in days:
        if(day=="all"):
            break
        print(day, " : ", len(df[df["start-days"]==day]["start-days"]))
        lists.append(len(df[df["start-days"]==day]["start-days"]))
    print("all : ",len(df["start-days"]))
    lists.append(len(df["start-days"]))
    return lists
```

```python
i= 0
fied_names_num(df[df["count"]>=i],names)
```
