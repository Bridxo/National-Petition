```python
import pandas as pd
```

```python
df = pd.read_csv("../petition_data_all.csv")
```

```python
def find_day(count):
    return count[:-3]
```

```python
df["start-days"] = df["start-days"].map(find_day)
df
```

```python
months = list(set(df["start-days"]))
months
```

```python
months_titles = []

for month in months:
    m_list = list(df[df["start-days"]==month]["title"])
    m_list = ' '.join(m_list)
    months_titles.append(m_list)
```

```python
len(months_titles)
```

```python
def fined_month_keyword(word):
    count_list = []
    for months_title in months_titles:
        count_list.append(months_title.count(word))
    return count_list
```

```python
te1 = fined_month_keyword("이명박")
te1
```

```python
te2 = fined_month_keyword("난민")
te2
```

```python
ms = []
for m in months:
    ms.append(m[3:4]+m[5:])
ms
```

```python
import matplotlib.font_manager as fm
```

```python
t_100 = ["난민", "폭행", "처벌", "국회의원", "대한항공"]
t_100_list = []

for t in t_100:
    t_100_list.append(fined_month_keyword(t))
```

```python
font_name = fm.FontProperties(fname = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf").get_name()
mpl.rc('font', family = font_name)

plt.subplots(1, figsize = (20,6))
for t_l,t_m in zip(t_100_list,t_100):
    plt.plot(months,t_l,label=t_m)
plt.grid()
#plt.legend()
plt.xlabel("Month")
plt.ylabel("Number of appearances")
plt.title("Graph 100+ suggestions every month")
plt.show()
```
